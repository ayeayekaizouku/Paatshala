### Computing the derived column day of the week

```python
def day_of_week_name(day_num):
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    return days[day_num]

df['dayOfWeek']= df['dayOfWeek'].map(day_of_week_name)
```

### Removing 2021

```python
df = df[df['Year'] != 2021]
```

### Year vs Likes

```python
grouped_data = df['Likes'].groupby(df['Year']).sum()
grouped_data = grouped_data.reset_index()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Year', y='Likes', data=grouped_data, color='purple')
plt.xlabel('Years')
plt.ylabel('Likes')
plt.title('Years vs Likes')
plt.show()
```

Covid seems to have increased the popularity of instagram, and online things in general, which has increased number of likes.

### Month vs Likes

```python
def month(day_num):
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    return months[day_num - 1]

df['Month'] = df['Month'].map(month)
```

```python
df
```

```python
grouped_data = df['Likes'].groupby(df['Month']).sum()
grouped_data = grouped_data.reset_index()
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.figure(figsize=(8, 6))
sns.barplot(x= 'Month',y='Likes', data=grouped_data,order= months_order, color='purple')
plt.xlabel('Months')
plt.ylabel('Likes')
plt.title('Months vs Likes')
plt.xticks(rotation=90)
plt.show()
```

Likes are generated more at the end of the year

### Quarter vs Likes

```python
grouped_data = df['Likes'].groupby(df['quarter']).sum()
grouped_data = grouped_data.reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='quarter', y='Likes', data=grouped_data, color='purple')
plt.xlabel('Quarter')
plt.ylabel('Likes')
plt.title('Quarter vs Likes')
plt.show()
```

3rd and 4th quarter generates more likes

### Hour vs Likes

```python
grouped_data = df['Likes'].groupby(df['Hour']).sum()
grouped_data = grouped_data.reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='Hour', y='Likes', data=grouped_data, color='purple')
plt.xlabel('Hour')
plt.ylabel('Likes')
plt.title('Hour vs Likes')
plt.show()
```

Likes take a dip when posted at noon.

### Time Of Day vs Likes

```python
grouped_data = df['Likes'].groupby(df['timeOfDay']).sum()
grouped_data = grouped_data.reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='timeOfDay', y='Likes', data=grouped_data, color='purple'
           , order= ['Early morning', 'Morning', 'Afternoon','Evening', 'Night', 'Late Night'])
plt.xlabel('Time of Day')
plt.ylabel('Likes')
plt.title('Time of Day vs Likes')
plt.show()
```

Overwhelmingly, posting at night generates more likes

### Day of Week vs Likes

```python
grouped_data = df['Likes'].groupby(df['dayOfWeek']).sum()
grouped_data = grouped_data.reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='dayOfWeek', y='Likes', data=grouped_data, color='purple'
           , order= ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.xlabel('Day of Week')
plt.ylabel('Likes')
plt.title('Day of Week vs Likes')
plt.show()
```
